```python

import pandas as pd

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames

# lo primero que hacemos es cargar el dataframe que vamos a usar para esta lección
df = pd.read_csv("HR RAW DATA transformado 02-06.csv") 
```

```python
# Los que se van, ¿tienen mayor salario?
df.groupby("attrition")["monthlyincome"].mean().round(2)
```

```python
sns.boxplot(data=df, x="attrition", y="monthlyincome")
plt.title("Distribución de salario según si se van o no")
plt.show()
```

```python
 # Relación entre salario y años en la empresa, separados por attrition
sns.scatterplot(data=df, x="yearsatcompany", y="monthlyincome", hue="attrition")
plt.title("Salario vs Años en la empresa según rotación")
plt.show()
```

```python
# Comparar la media de satisfacción laboral
df.groupby("attrition")["jobsatisfaction"].mean().round(2)
```

```python
sns.countplot(x="jobsatisfaction", hue="attrition", data=df)

plt.title("Satisfacción laboral por rotación")
plt.xlabel("Nivel de satisfacción laboral (1 = baja, 4 = alta)")
plt.ylabel("Número de empleados")
plt.legend(title="¿Se fue de la empresa?")
plt.show()
```

```python
sns.boxplot(x="attrition", y="yearssincelastpromotion", data=df)

plt.title("Años desde la última promoción por rotación")
plt.xlabel("¿El empleado se fue?")
plt.ylabel("Años desde la última promoción")
plt.show()
```

```python
sns.boxplot(x="attrition", y="worklifebalance", data=df)
plt.title("Equilibrio vida-trabajo por rotación")
plt.xlabel("¿El empleado se fue?")
plt.ylabel("Equilibrio vida-trabajo (1 = bajo, 4 = alto)")
plt.show()
```

```python
# Worklifebalance no nos aporta nada, porque tienen una distribución similar.
# El equilibrio vida-trabajo no parece ser un factor determinante.
```
